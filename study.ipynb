```python
# НАСЛЕДОВАНИЕ - создание нового класса (класс наследник), который унаследует все поля и методы из основого класса 
class Person:
    name="Inan" 
    age=10
    def set(self, name, age):
        self.name=name
        self.age=age

class Student(Person):
    course=1
    
igor=Student() # Нужно объявить объект
igor.set("Игорь",19) # Потом его инициализировать
print(igor.name)
print(igor.course)
igor.cource=2
print(igor.cource)
print('-------------------------')
```

```python
# ИНКАПСУЛЯЦИЯ - ограничение доступа к каким-либо полям или методам(защита данных)
# поле name доступно как в Person, так и в Student, так и в объектах
# аналогично метод set доступен в двух классах
# добавление '_'(одно нижнее подчёркивание): (_name) или (_set) ФОРМАЛЬНО ограничивет использование поля или метода
# это ограничение лишь формально для других программистов.

# использование '__' (два нижних подчёркивание): (__name) или (__set) возвращет ошибку при использовании вне класса
class Person:
    name="Inan" 
    age=10
    def __set(self, name, age):
        self.name=name
        self.age=age

class Student(Person):
    course=1

igor=Student()
igor.name='Игорь'
print(igor.name)

try:
    igor.set('ИГОРЁК',34)
except:
    print('ERROR1')

    
print('----------------')    


try:
    igor.__set('ИГОРЁК')
except:
    print('ERROR2')
 
    
# однако можно сделать так:
igor._Person__set('ИГОРЁК',34)
print(igor.name+' '+str(igor.age))
```

```python
class Hero():
	# в классах функции называются методами
	def __init__(self,name,level,race):
		# Инициализая героя
		self.name=name
		self.level=level 
		self.race=race
		self.health=100
	def show_hero(self):
		#Print all parameters of this hero
		discription =(self.name+' Level is: '+str(self.level) + ' Race is: ' + self.race + ' Helth is:'+str(self.health)).title()
		print(discription)
	def level_up(self):
		#Upgrade level  of Hero
		self.level+=1
	def move(self):
		#start moving Hero
		print('Hero '+self.name+' start moving...')
# What is create object?
# инициализация объекта при наличии метода __init__:
myhero1=Hero('vurdalak',5,'Orc')
myhero2=Hero('Aleksandr',4,'Human') 

myhero1.show_hero()
myhero2.move()
myhero1.level_up()
myhero1.show_hero()
myhero1.health=70
myhero1.show_hero()
```

```python
class SuperHero(Hero):
    #__init__ запускается при создании супергероя
    def __init__(self,name,level,race,magiclevel):
        #чтобы не инициализировать все переменные два раза, можно сделать так:
        #перекидываем параметры, которые мы не хотим повторно инициализировать,т.к. они есть в родительском классе:
        super().__init__(name,level,race)
        #инициализируем параметры, которых нет в родительском классе:
        self.magiclevel=magiclevel
        self.magic=100
    def makemagic(self):
        self.magic-=10
    # как теперь расширить функцию show_hero, добавлением праметра magiclevel 
    def show_hero(self):
        discription =(self.name+' Level is: '+str(self.level) + ' Race is: ' + self.race + ' Helth is:'+str(self.health)+' Magic is: '+str(self.magic)).title()
        print(discription)
```

```python
# Подключаем модуль threading для работы с потоками
import threading
# и модули requests, shutil, os для скачивания файлов из сети
import requests, shutil, os

# Функция скачивания файла которую мы будем выполнять
# в новом потоке на входе номер потока и URL скачиваемого файла
def down(nomer, url):
    # Выводим на экран сообщение о старте нового потока
    print(str(nomer)+' поток ('+url+')\n')
    # Получаем имя файла из URL
    (dirname, filename) = os.path.split(url)
    # Запускаем скачивание через GET запрос
    r = requests.get(url, stream=True)
    # Если код ответа сервера 200 (сервер доступен)
    if r.status_code == 200:
        # Открываем файл в режиме побайтовой записи
        with open(filename, 'wb') as f:
            r.raw.decode_content = True
            # Созхраняем скачиваемый поток в файл
            shutil.copyfileobj(r.raw, f)
        # Выводим сообщение что поток закончил загрузку
        print(str(nomer) + " поток закончил загрузку\n")
        

# Список URL которые будем скачивать
urls = ["http://www.irs.gov/pub/irs-pdf/f1040.pdf",
            "http://www.irs.gov/pub/irs-pdf/f1040a.pdf",
            "http://www.irs.gov/pub/irs-pdf/f1040ez.pdf",
            "http://www.irs.gov/pub/irs-pdf/f1040es.pdf",
            "http://www.irs.gov/pub/irs-pdf/f1040sb.pdf"]

# Цикл который создаст столько потоков сколько файлов нужно скачать
for nomer, url in enumerate(urls):
    # Запускаем поток указав целью функцию down и ее аргументы
    # к nomer прибавляем 1 чтобы номер потока начинался не с нуля
    threading.Thread(target=down, args = [nomer+1, url]).start()

# Выводим сообщение демонстрирующее что основная программа
# продолжает работать отдельно от созданных потоков
print('\nЕще не завершились потоки а программа уже продолжается\n')
```

```python
import time

class parsing:
    # параметры файла cfg
    param_name = [1, str]
    param_phase = [2, str]
    param_dimension = [4, str]
    param_a = [5, float]
    param_b = [6, float]
    cfg_dict = {}

    def __init__(self, x, dat_name, cfg_name):
        self.stolbec = x
        self.dat_name = dat_name
        self.cfg_name = cfg_name
        self.list_of_lists = []
        self.cfg_list = []

    def time(self):
        f = open(self.dat_name, 'r')  # Файл с данными
        x = f.read().split('\n')[:-1]
        for line in x:
            my_list = [int(elt) for elt in line.split(',')]
            self.list_of_lists.append(my_list)
        f.close()

    def parse(self):
        f = open(self.dat_name, 'r')  # Файл с данными
        x = f.read().split('\n')[:-1]
        for line in x:
            my_list = [int(elt) for elt in line.split(',')]
            self.list_of_lists.append(my_list)
        f.close()

    def get_stolbec(self):
        for i in self.list_of_lists:
            yield (i[self.stolbec])

    def get_state(self):
        for i in self.list_of_lists:
            yield i

    def read_cfg(self):
        f = open(self.cfg_name, 'r')
        x = f.read().split('\n')[2:8]
        n = 0
        for line in x:
            cfg_list = []
            n += 1
            for i in line.split(','):
                cfg_list.append(i)
            cfg_name = cfg_list[self.param_name[0]]
            cfg_phase = cfg_list[self.param_phase[0]]
            cfg_dimension = cfg_list[self.param_dimension[0]]
            cfg_a = float(cfg_list[self.param_a[0]])
            cfg_b = float(cfg_list[self.param_b[0]])
            self.cfg_dict[n] = [cfg_name, cfg_phase, cfg_dimension, cfg_a, cfg_b]
        f.close()

import math
Time1=[]
CurA_kz=[]
CurB_kz=[]
CurC_kz=[]
VolA_kz=[]
VolB_kz=[]
VolC_kz=[]
CurA_nm=[]
CurB_nm=[]
CurC_nm=[]
VolA_nm=[]
VolB_nm=[]
VolC_nm=[]
S=Time1
S_N=S

def phase(a1,a2,a3,name):
    print('start',name)
    file = parsing(a1, "M1_ABCGO.dat", "M1_ABCGO.cfg")
    file.parse()
    Time = [i / 1e6 for i in file.get_stolbec()]
    Time1.extend(i for i in Time)
    file.read_cfg()

    # вывод тока
    file = parsing(a2, "M1_ABCGO.dat", "M1_ABCGO.cfg")
    file.parse()
    Cur = [(i * file.cfg_dict[1][3] + file.cfg_dict[1][4]) for i in file.get_stolbec()]
    if name=='phA_nm':
        CurA_nm.extend(i for i in Cur)
    if name=='phB_nm':
        CurB_nm.extend(i for i in Cur)
    if name=='phC_nm':
        CurC_nm.extend(i for i in Cur)
    if name=='phA_kz':
        CurA_kz.extend(i for i in Cur)
    if name=='phB_kz':
        CurB_kz.extend(i for i in Cur)
    if name=='phC_kz':
        CurC_kz.extend(i for i in Cur)
    file = parsing(a3, "M1_ABCGO.dat", "M1_ABCGO.cfg")
    file.parse()
    Vol = [(i * file.cfg_dict[5][3] + file.cfg_dict[5][4]) for i in file.get_stolbec()]
    if name=='phA_nm':
        VolA_nm.extend(i for i in Cur)
    if name=='phB_nm':
        VolB_nm.extend(i for i in Cur)
    if name=='phC_nm':
        VolC_nm.extend(i for i in Cur)
    if name=='phA_kz':
        VolA_kz.extend(i for i in Vol)
    if name=='phB_kz':
        VolB_kz.extend(i for i in Vol)
    if name=='phC_kz':
        VolC_kz.extend(i for i in Vol)
    S= [a * b for a in Cur for b in Vol]
    time.sleep(0.1)
    print('finish',name)

class Profiler_object():
    def __enter__(self):
        self._startTime=time.time()
    def __exit__(self, exc_type, exc_val, exc_tb):
        self._endTime=time.time()
        print('time:',self._endTime-self._startTime)





import sqlite3 as db


meas_id_Ia=1
meas_id_Ib=2
meas_id_Ic=3
meas_id_Ua=4
meas_id_Ub=5
meas_id_Uc=6



norm_rezh_id=1
KZ_rezh_id=2

def Measurement(Id, Time, value, meas_id, rezh_id):
    c = db.connect("ЛР №5.db")
    cursor = c.cursor()
    for i in range(0, len(value)):
        if (Id == 0):
            cursor.execute("""INSERT OR REPLACE INTO datapoints
                        (Id,Time,value,meas_id,rezh_id,S,S_N) VALUES("%s","%s","%s","%s","%s","%s","%s")"""
                           % (i + Id * len(value), Time[i], value[i], meas_id, rezh_id, S[i], S_N[i]))
        else:
            cursor.execute("""INSERT OR REPLACE INTO datapoints
                        (Id,Time,value,meas_id,rezh_id) VALUES("%s","%s","%s","%s","%s")"""
                           % (i + Id * len(value), Time[i], value[i], meas_id, rezh_id))
    c.commit()


with Profiler_object() as P:
    phaseA_nm = phase(1, 2, 5, 'phA_nm')
    phaseB_nm = phase(1, 3, 6, 'phB_nm')
    phaseC_nm = phase(1, 4, 7, 'phC_kz')
    phaseA_kz = phase(1, 2, 5, 'phA_kz')
    phaseB_kz = phase(1, 3, 6, 'phB_kz')
    phaseC_kz = phase(1, 4, 7, 'phC_kz')
    #Заполнение данных нормального режима в БД
    Measurement(0,Time1,CurA_nm,meas_id_Ia,norm_rezh_id)
    Measurement(1,Time1,CurB_nm,meas_id_Ib,norm_rezh_id)
    Measurement(2,Time1,CurC_nm,meas_id_Ic,norm_rezh_id)
    Measurement(3,Time1,VolA_nm,meas_id_Ua,norm_rezh_id)
    Measurement(4,Time1,VolB_nm,meas_id_Ub,norm_rezh_id)
    Measurement(5,Time1,VolC_nm,meas_id_Uc,norm_rezh_id)

    #Заполнение данных режима КЗ в БД
    Measurement(6, Time1, CurA_kz, meas_id_Ia, KZ_rezh_id)
    Measurement(7, Time1, CurB_kz, meas_id_Ib, KZ_rezh_id)
    Measurement(8, Time1, CurC_kz, meas_id_Ic, KZ_rezh_id)
    Measurement(9, Time1, VolA_kz, meas_id_Ua, KZ_rezh_id)
    Measurement(10, Time1, VolB_kz, meas_id_Ub, KZ_rezh_id)
    Measurement(11, Time1, VolC_kz, meas_id_Uc, KZ_rezh_id)
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

recipe = ["375 g flour",
          "75 g sugar",
          "250 g butter",
          "300 g berries"]

data = [float(x.split()[0]) for x in recipe]
ingredients = [x.split()[-1] for x in recipe]


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          title="диагнозы",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("DDD")

plt.show()
fig.savefig('f5.png',dpi = 200);
```

```python
def loc_decorator(func):
    def wrap():
        print(f'Calling func {func}')
        func()
        print(f'Func {func} finished its works')
    return wrap
```

```python
def hello():
    print('Hello, world!')
```

```python
wrapped_dy_loger = loc_decorator(hello)
wrapped_dy_loger()
```

```python
from timeit import default_timer as timer
import math
import time
```

```python
# С помощью декоратора замерим длительность функции
def measure_time(func):
    def inner(*args,**kwargs):
        print(f'Calling function {func.__name__}:\n')
        start = timer()
        
        func(*args,**kwargs)
        
        end = timer()
        print(f'\nFunction {func.__name__} finished its works.')
        print(f'Function {func.__name__} took {end-start} sec for execution.')
    return inner
```

```python
@measure_time
def factorial(num):
    time.sleep(3)
    print(math.factorial(num))
```

```python
factorial(150)
```

```python
def benchmark(func):
    import time
    
    def wrapper():
        start = time.time()
        func()
        end = time.time()
        print(f'[*] Время выполнения: {end-start} секунд.')
    return wrapper

@benchmark
def fetch_webpage():
    import requests
    webpage = requests.get('https://google.com')

fetch_webpage()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
x = np.random.randint(low=1, high=11, size=50)
y = x + np.random.randint(1, 5, size=x.size)
data = np.column_stack((x, y))
 
fig, (ax1, ax2) = plt.subplots(
    nrows=1, ncols=2,
    figsize=(8, 4)
)
 
ax1.scatter(x=x, y=y, marker='o', c='r', edgecolor='k')
ax1.set_title('Scatter: $x$ versus $y$')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
 
ax2.hist(
    data, bins=np.arange(data.min(), data.max()),
    label=('x', 'y')
)
 
ax2.legend(loc=(0.65, 0.8))
ax2.set_title('Frequencies of $x$ and $y$')
ax2.yaxis.tick_right()
 
plt.show()
```
