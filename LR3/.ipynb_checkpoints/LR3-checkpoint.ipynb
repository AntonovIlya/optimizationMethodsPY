<p>Декоратор для измерения времени работы алгоритма</p>

```python
def timeMeasurement(function):
    def wrapper():
        start = time.time()
        function()
        end = time.time()
        global run_time 
        run_time = end-start
        print(f'\033[95m[*] run time function({function.__name__}): {run_time:1.5f} s \033[0m')
    return wrapper
```

<p>Функция инициализации начальной популяции</p>

```python
def init():
    population = []
    for i in range(n):
        genotype = []
        for i in range(len(word5)):
            genotype.append(random.choice(ABC4))
        population.append(genotype)
    return population
```

<p>Функция для определения фенотипа хромосомы</p>

```python
def phenotype_mod_one(genotype):
    summ = 0
    for i in range(len(genotype)):
        summ += abs(ABC4.index(word5[i]) - ABC4.index(genotype[i]))
    return summ
```

<p>Функция для определения среднего фенотипа по популяции</p>

```python
def average_phenotype(population):
    summ = 0
    for i in population:
        summ += phenotype_mod_one(i)
    return (summ / len(population))
```

<p>Оператор рулеточной селекции</p>

```python
def roulette_selection(population):
    global flag
    total_population = []
    len_total_population = math.floor(s * len(population))
    while (len(total_population) < len_total_population) and flag:
        p = []
        summ = 0
        for i in population:
            if (phenotype_mod_one(i) == 0):
                flag = False
                #print(i)
                break
            p.append(1 /phenotype_mod_one(i))
        r = random.uniform(0, sum(p))
        for i in range(len(p)):
            summ += p[i]
            if (r < summ):
                total_population.append(population[i])
                population.remove(population[i])
                break
    return total_population
```

<p>Оператор турнирной селекции</p>

```python
def tournament_selection(population):
    global flag
    flag = True
    for i in population:
        if (phenotype_mod_one(i) == 0):
            #print(i)
            flag = False
    total_population = []
    len_total_population = math.floor(s * len(population))
    while (len(total_population) < len_total_population) and flag:
        tournament = {}
        for i in range(t):
            new_individ = random.choice(population)
            while phenotype_mod_one(new_individ) in tournament.keys():
                new_individ = random.choice(population)
            tournament[phenotype_mod_one(new_individ)] = new_individ
        total_population.append(tournament[min(tournament.keys())])
        population.remove(tournament[min(tournament.keys())])
    return total_population
```

<p>Функция скрещевиния в одной точке</p>

```python
def cross_one_point(parent_a, parent_b):
    point = random.choice(range(1,len(parent_a)))
    child_a = parent_a[:point] + parent_b[point:]
    child_b = parent_b[:point] + parent_a[point:]
    return [child_a, child_b]
```

<p>Функция скрещивания в двух точках</p>

```python
def cross_two_point(parent_a, parent_b):
    point_one = random.choice(range(1,len(parent_a)))
    point_two = random.choice(range(1,len(parent_a)))
    while point_one == point_two:
        point_two = random.choice(range(1,len(parent_a)))
    if (point_one < point_two):
        child_a = parent_a[:point_one] + parent_b[point_one:point_two] + parent_a[point_two:]
        child_b = parent_b[:point_one] + parent_a[point_one:point_two] + parent_b[point_two:]
    else:
        child_a = parent_a[:point_two] + parent_b[point_two:point_one] + parent_a[point_one:]
        child_b = parent_b[:point_two] + parent_a[point_two:point_one] + parent_b[point_one:]
    return [child_a, child_b]
```

<p>Функция скрещивания, родительские пары выбираются случайно, новая популяция состоит только из новых особей.</p>

```python
def crossing(population):
    k = 0
    global flag2, flag3
    flag2 = True
    flag3 = True
    total_population = []
    if len(population) == 0:
        flag3 = False
        print("Популяция умерла")
    if (len(remove_equal_genotype(population)) == 1):
        print("Популяция выродилась")
        flag2 = False
    else:
        while k <= n and flag2 and flag3:
            count = 0
            parent_a = random.choice(population)
            parent_b = random.choice(population)
            while parent_b == parent_a:
                parent_b = random.choice(population)
            if (crossing_probability > random.random()):
                total_population += cross_one_point(parent_a, parent_b)
            k = k + 2 
    return total_population
```

<p>Функция мутации</p>

```python
def mutation(population):
    for i in range(len(population)):
        for j in range(len(population[i])):
            if (mutation_probability > random.random()):
                population[i][j] = random.choice(ABC4)
```

<p>Основной алогритм для первой задачи. Критерии остановки: вырождении популяции, ограничение числа поколений, смерть популяции, нахождении глобального минимума функции приспособленности. </p>

```python
@timeMeasurement
def main():
    population = init()
    global flag, flag2, flag3
    flag2 = True
    flag3 = True
    flag = True
    count = 0
    while flag and flag2 and flag3 and count < 200:
        #print(average_phenotype(population))
        population = crossing(population)
        mutation(population)
        population = tournament_selection(population)
        count += 1
```

```python
def remove_equal_genotype(population):
    new_population = []
    for i in population:
        if (i not in new_population):
            new_population.append(i)
    return new_population
```
