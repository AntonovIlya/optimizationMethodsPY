<div align="justify">Муравьиные алгоритмы - семейство эффективных приближённых алгоритмов для решения различных сложных оптимизационных задач. В данной работе будет реализована его базовая вариация, алгоритм элитной муравьиной системы и алгоритм муравьиной колонии. </div>
<p>Импорт библиотек и загрузка матрицы смежности для решения задачи коммивояжера:</p>

```python
import numpy
import random
import matplotlib.pyplot as plt
import time
#array_numpy = numpy.load("Task2.npy")
numpy.set_printoptions(precision=1, suppress = True)
#print(array_numpy)
shape = (20, 20)
array_numpy = numpy.zeros(shape)
point_list = numpy.array([i for i in range(1, 21)])
for i in range(20):
    for j in range(20):
        if (i == j):
            array_numpy[i][j] = 0
        elif (i == 0):
            array_numpy[i][j] = point_list[j]
        elif (j == 0):
            array_numpy[i][j] = point_list[i]
        else:
            array_numpy[i][j] = random.uniform(1, 100)
            array_numpy[j][i] = array_numpy[i][j]
print(array_numpy)
```

<p>Реализуем базовый алгоритм, для этого зададим начальные значения параметров:</p>

```python
a = 1
b = 1
t0 = 1
ants = 60
steam = 0.9
Q = 10
number_iter = 5
pheromones = numpy.array(array_numpy, dtype = float)
```

<p>Функция, возвращающая расстояние между двумя городами:</p>

```python
def lenght(a, b):
    return array_numpy[list(array_numpy[0]).index(a), list(array_numpy[0]).index(b)]
```

<p>Функция, возвращающая длину пути полученной последовательности городов:</p>

```python
def lenght_way(array):
    s = 0
    for i in range(len(array)):
        s += lenght(array[i-1], array[i])
    return s
```

<p>Декоратор для измерения времени работы алгоритма:</p>

```python
def timeMeasurement(function):
    def wrapper():
        start = time.time()
        function()
        end = time.time()
        global run_time 
        run_time = end-start
        #print(f'\033[95m[*] run time function({function.__name__}): {run_time:1.5f} s \033[0m')
    return wrapper
```

<p>Функция перебора возможных путей (184 = минимальная длина пути):</p>

```python
@timeMeasurement
def test():
    best = []
    list_points =[array_numpy[0,i] for i in range(1,10)]
    for i in range(200000):
        numpy.random.shuffle(list_points)
        best.append(lenght_way(list_points))
    list_ways = list(set(best))
    list_ways.sort()
    print(f'The first 10 minimum ways: {list_ways[:10]}')
    print('lenght = 184:')
    for i in range(200000):
        numpy.random.shuffle(list_points)
        if lenght_way(list_points) == 184:
            print(list_points)
```

```python
test()
```

<p>Ранжированный список путей: 184, 197, 203, 207, 216. Следовательно критерием остановки для всех реализаций будет значение найденного пути меньше 216.
<p>Как видно, начиная практически с любой начальной вершины можно получить минимальную длину пути для данной задачи с 9 городами</p>

<p>Функция, инициализирующая начальные значения феромонов в рёбрах:</p>

```python
def init_pheromones():
    for i in range(1,10):
        for j in range(1,10):
            pheromones[i,j] = t0
```

<p>Функция, возвращающая феромон ребра:</p>

```python
def pheromone(i, j):
    return pheromones[list(array_numpy[0]).index(i), list(array_numpy[0]).index(j)]
```

<p>Функция, возвращающая значение вероятности перехода из текущей точки в последующую на основе списка возможных вариантов:</p>

<p>формула</p>

```python
def P(current_point, next_point, access_list):
    up = (pheromone(current_point, next_point) ** a) * ((1 / lenght(current_point, next_point)) ** b)
    down = 0
    for i in access_list:
        down += (pheromone(current_point, i) ** a) * ((1 / lenght(current_point, i)) ** b)
    return  up / down
```

<p>Функция, возвращающая номер следующего города по следующему принципу. На вход подаётся список вероятностей перехода в разные города, чем больше вероятность, тем больше область в которую может попасть рандомное число и произвести переход.</p>

```python
def next_point_index(P_list):
    s = 0
    r = random.random()
    for i in range(len(P_list)):
        s += P_list[i]
        if (r < s): 
            return i
```

<p>Функция, обновляющая значения феромонов на рёбрах, но основе списка городов, полученного на входе:</p>

```python
def updatePheromones(way):
    for i in range(1,10):
        for j in range(1,10):
            pheromones[i,j] = pheromones[i,j] * steam
    for i in range(len(way)):
        index_one = list(array_numpy[0]).index(way[i-1])
        index_two = list(array_numpy[0]).index(way[i])
        pheromones[index_one, index_two] += Q / lenght_way(way)
        pheromones[index_two, index_one] = pheromones[index_one, index_two]
```

<p>Модификация функции обновления феромонов на рёбрах, отличиющаяся наличием ограничений на значения феромонов снизу и сверху:</p>

```python
def updatePheromones_mod(way):
    for i in range(1,10):
        for j in range(1,10):
            if pheromones[i,j] * steam < tmin :
                pheromones[i,j] = tmin
            else:
                pheromones[i,j] = pheromones[i,j] * steam
    for i in range(len(way)):
        index_one = list(array_numpy[0]).index(way[i-1])
        index_two = list(array_numpy[0]).index(way[i])
        if (pheromones[index_one, index_two] + Q / lenght_way(way) > tmax):
            pheromones[index_one, index_two] = tmax
            pheromones[index_two, index_one] = tmax
        else: 
            pheromones[index_one, index_two] += Q / lenght_way(way)
            pheromones[index_two, index_one] = pheromones[index_one, index_two]
```

<p>Базовая реализация алгоритма:</p>

```python
@timeMeasurement
def main():
    global lenght_best_way
    init_pheromones()
    lenght_best_way = 1000
    list_of_points =[array_numpy[0,i] for i in range(1,10)]
    initial_point = random.choice(list_of_points)
    #while lenght_best_way > 220:
    for i in range(5):
        P_list = []
        ways_list = []
        for k in range(ants):
            list_of_points =[array_numpy[0,i] for i in range(1,10)]
            current_point = initial_point
            way =[current_point]
            list_of_points.remove(current_point)
            while len(list_of_points) != 1:
                for i in list_of_points:
                    P_list.append(P(current_point, i, list_of_points))
                current_point = list_of_points[next_point_index(P_list)]
                way.append(current_point)
                P_list.clear()
                list_of_points.remove(current_point)
            way.append(list_of_points[0])
            if (lenght_way(way) < lenght_best_way): 
                best_way = way
                lenght_best_way = lenght_way(best_way)
            ways_list.append(way)
        for i in ways_list:
            updatePheromones(i) 
    #print(f'\033[94m best way on iteration: {best_way}, lenght: {lenght_best_way} \033[0m')
```

```python
for i in range(5):
    main()
```

<h2><center>Модель элитной муравьиной системы</center></h2>
<p>Обновление феромона производится только для лучшего найденного пути</p>
<p>Добавляются ограничения на концентрацию феромона на ребре</p>

```python
@timeMeasurement
def main_mod():
    global lenght_best_way
    lenght_best_way = 1000
    init_pheromones()
    list_of_points =[array_numpy[0,i] for i in range(1,10)]
    initial_point = random.choice(list_of_points)
    while lenght_best_way > 216:
        P_list = []
        for k in range(ants):
            list_of_points =[array_numpy[0,i] for i in range(1,10)]
            current_point = initial_point
            way =[current_point]
            list_of_points.remove(current_point)
            while len(list_of_points) != 1:
                for i in list_of_points:
                    P_list.append(P(current_point, i, list_of_points))
                current_point = list_of_points[next_point_index(P_list)]
                way.append(current_point)
                P_list.clear()
                list_of_points.remove(current_point)
            way.append(list_of_points[0])
            if (lenght_way(way) < lenght_best_way): 
                best_way = way
                lenght_best_way = lenght_way(best_way)
                updatePheromones_mod(best_way) 
    print(f'\033[94m best way on iteration: {best_way}, lenght: {lenght_best_way} \033[0m')
```

```python
tmin = 1
tmax = 5
for i in range(5):
    main_mod()
```

<h2><center>Алгоритм муравьиной колонии</center></h2>
<p>Локальное обновление фепомона после прохождения по ребру</p>
<p>Глобальное обновление для лучшего пути на итерации</p>
<p>Выбор следующего ребра псевдо-случайным образом</p>
<p>Функция, обоновляющая феромон ребра:</p>

```python
def update_pheromone(a, b):
    index_one = list(array_numpy[0]).index(a)
    index_two = list(array_numpy[0]).index(b)
    #print(index_one, index_two)
    pheromones[index_one, index_two] = (1 - ph) * pheromones[index_one, index_two] + ph * t0
    pheromones[index_two, index_one] = (1 - ph) * pheromones[index_two, index_one] + ph * t0
```

<p>Функция, возвращающая номер следующей точки с максимальной концентрацией феромона на ребре:</p>

```python
def max_pheromone(current_point, list_of_points):
    max_pheromone = 0
    index_one = list(array_numpy[0]).index(current_point)
    for i in list_of_points:
        index_two = list(array_numpy[0]).index(i)
        if pheromones[index_one, index_two] > max_pheromone:
            max_pheromone = pheromones[index_one, index_two]
            next_point = i
    return next_point
```

```python
@timeMeasurement
def main_mod_two():
    global lenght_best_way
    init_pheromones()
    lenght_best_way = 1000
    list_of_points =[array_numpy[0,i] for i in range(1,10)]
    initial_point = random.choice(list_of_points)
    while lenght_best_way > 216:
        P_list = []
        ways_list = []
        for k in range(ants):
            list_of_points =[array_numpy[0,i] for i in range(1,10)]
            current_point = initial_point
            way =[current_point]
            list_of_points.remove(current_point)
            while len(list_of_points) != 1:
                for i in list_of_points:
                    P_list.append(P(current_point, i, list_of_points))
                if (random.random() < q0):
                    next_point = max_pheromone(current_point, list_of_points)
                else:
                    next_point = list_of_points[next_point_index(P_list)]
                update_pheromone(current_point, next_point)
                current_point = next_point
                way.append(current_point)
                P_list.clear()
                list_of_points.remove(current_point)
            way.append(list_of_points[0])
            if (lenght_way(way) < lenght_best_way): 
                best_way = way
                lenght_best_way = lenght_way(best_way)
            ways_list.append(way)
        updatePheromones(best_way) 
    print(f'\033[94m best way on iteration: {best_way}, lenght: {lenght_best_way} \033[0m')
```

```python
ph = 0.9
q0 = 0.5
for i in range(5):
    main_mod_two()
```
